**2. Age 변수 결측값 채우기**

-전체 나이 평균 / 중앙값으로 채우기

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
```

-Pclass 별 나이 평균으로 채우기 

```python
data['Age'] = data['Age'].fillna(
    data.groupby(['Pclass'])['Age'].transform('median')
)
```

-Sex & Pclass 나이 평균으로 채우기

```python
data['Age'] = data['Age'].fillna(
    data.groupby(['Sex', 'Pclass'])['Age'].transform('median')
)
```

-Name에서 호칭 추출해서 중앙값으로 채우기

```python
data['Title'] = data['Name'].str.extract(' ([A-Za-z]+)\.')
# ex) Braund, Mr. Owen Harris  ->  ' ': 공백 하나, [A-Za-z]+: 알파벳 1글자 이상, (): 이 안에 있는 것을 추출, \.: 마침표

data['Age'] = data.groupby('Title')['Age'].transform(
    lambda x: x.fillna(x.median())
)
```
