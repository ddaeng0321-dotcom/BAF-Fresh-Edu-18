# **기온 결측치 채우기**
![설명](이슬점공식.png)

- Dewpoint = 이슬점
- T = 기온
- RH = humidity

```python
data_train.isnull().sum()
# 기온, 풍속, 계절에 결측치가 존재함.
# 기상청 데이터에서 해당 날짜의 습도와 이슬점 온도가 같은 것을 확인하 결측치를 대체할 것임.
```

```python
data_train['Date'] = pd.to_datetime(data_train['Date'], dayfirst=True)
# Date 변수 datetime으로 바꾸기
# dayfirst=True : 01/12/2017 → 12월 1일로 인식하도록 함
```

```python
data_train['Date'].min(), data_train['Date'].max()
# train 데이터셋에 Date 변수는 2017년12월1일 부터 2018년10월31
```

```python
asos = pd.read_csv("OBS_ASOS_TIM_20260206163458.csv", encoding='cp949')
asos.head()
```

```python
asos['datetime'] = pd.to_datetime(asos['일시'])
# 기상청 datetime 만들기
```

```python
data_train['datetime'] = data_train['Date'] + \
                         pd.to_timedelta(data_train['Hour'], unit='h')
# 따릉이 datetime 만들기
```

```python
merged = data_train.merge(asos, on='datetime', how='left')
print(merged.columns)
```

```python
data_train['Temperature(째C)'] = data_train['Temperature(째C)'].fillna(merged['기온(°C)'])
data_train['Wind speed (m/s)'] = data_train['Wind speed (m/s)'].fillna(merged['풍속(m/s)'])
```

```python
data_train.isnull().sum()
# 아직 기온, 풍속, 계절에 결측치 존재
```

```python
asos['datetime'].min(), asos['datetime'].max()
# 기상청 데이터와 train 데이터셋의 시간범위가 일치하지 않아 결측값이 남아있다.
```

```python
data_train['Temperature(째C)'] = \
data_train['Temperature(째C)'].interpolate()

data_train['Wind speed (m/s)'] = \
data_train['Wind speed (m/s)'].interpolate()
# 시계열 데이터의 연속성을 고려하여 인접 시간대 값을 기반으로 선형 보간(interpolation)을 적용해 최종적으로 결측치를 처리하였다.
# 선형보간: 결측값 앞뒤에 있는 실제 값 사이를 직선으로 이어서 중간값을 추정해 채우는 방법
```

```python
def get_season(month):
    if month in [3,4,5]:
        return 'Spring'
    elif month in [6,7,8]:
        return 'Summer'
    elif month in [9,10,11]:
        return 'Autumn'
    else:
        return 'Winter'
# 월별로 계절 정하는 함수 만들기

data_train['Seasons'] = data_train['Date'].dt.month.apply(get_season)
# Date에서 월만 뽑아서 get_season 함수 이용하여 계절로 바꾸기
```

```python
data_train.isnull().sum()
# 결측치가 존재하지 않는다! 
```

# **모델링 하기**

```python
data_train = pd.get_dummies(data_train, columns=['Seasons','Holiday','Functioning Day'], drop_first=True)
data_test = pd.get_dummies(data_test, columns=['Seasons','Holiday','Functioning Day'], drop_first=True)
# 문자열은 모델이 못 읽으므로 범주형 변수 인코딩하기
```

```python
data_test = data_test.reindex(columns=data_train.columns, fill_value=0)
# train/test 데이터 컬럼 맞추기
```

```python
X = data_train.drop(['Rented Bike Count','Date','datetime'], axis=1)
y = data_train['Rented Bike Count']
# X,y 나누기
```

```python
X_test = data_test.drop(['Rented Bike Count','Date','datetime'], axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(X, y)
# 랜덤포레스트 회귀모델로 학습시키기
# 따릉이 대여량은 단순히 한 변수로 결정되지 않고 시간대(Hour), 기온, 강수량, 계절, 휴일 여부 등이 복합적으로 작용하는 비선형 패턴을 가진다.
# 이러한 특성은 선형회귀보다 변수 간 복잡한 관계를 가진 비선형 구조를 잘 학습하는 랜덤포레스트 모델이 더 적합하다.
# 랜덤포레스트는 여러 개의 의사결정나무를 생성한 뒤, 각 나무의 예측 결과를 평균내어 최종 값을 예측하는 앙상블 기반 회귀 모델이므로 단일 모델보다 예측 안정성이 높고 과적합을 줄일 수 있는 장점이 있다.
```

```python
pred_rf = rf.predict(X_test)
# test 데이터로 예측하기
```

```python
submission = pd.DataFrame({
    'Rented Bike Count': pred_rf
})
submission.to_csv('submission_predict_data.csv', index=False)
```
